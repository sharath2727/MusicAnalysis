```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from matplotlib.pyplot import specgram
```

```python
def plot_sine(freqs, samp_rate):
    
    samp_duration = 1/samp_rate
    t = np.arange(0, 1, samp_duration)
    
    y = 0
    
    for freq in freqs:
        y += np.sin(2*np.pi*freq*t)
        
    plt.plot(t, y)
    plt.grid()
    plt.show()
        
    return y 

def fourier_analysis(time_func, samp_rate):
    
    k = np.arange(len(time_func))
    T = len(time_func)/samp_rate
    frq = k/T 
    frq = frq[range(int(len(time_func)/2))] # one side frequency range

    Y = np.fft.fft(time_func)/len(time_func) 
    Y = Y[range(int(len(time_func)/2))]
 
    plt.plot(frq,abs(Y),'r') # plotting the spectrum
    plt.xlabel('Freq (Hz)')
    plt.ylabel('|Amplitude|')
    plt.title('Fourier')
    plt.show()
    
def plot_spectrogram(time_func, samp_rate):
    specgram(time_func, Fs=samp_rate)
    plt.xlabel('Time')
    plt.ylabel('Freq (Hz)')
    plt.title('Spectrogram')
    plt.show()
    
```

```python
if __name__ == '__main__':
    samp_rate = 500
    time_func = plot_sine([10, 20], samp_rate)
    fourier_analysis(time_func, samp_rate)
    plot_spectrogram(time_func, samp_rate)
    # plt.show()
```
